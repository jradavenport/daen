# The SETI Ellipsoid

Let's code up the echo [ellipsoid](https://en.wikipedia.org/wiki/Ellipsoid) that expands with time for an event

This is based on the presentation in [Lemarchand (1994)](https://ui.adsabs.harvard.edu/abs/1994Ap%26SS.214..209L/) discussing SETI based around SN 1987A, but the concept is identical to the light echo mapping used to study dust structure around SN 1987A (e.g. [this talk](https://cxc.harvard.edu/cdo/snr09/pres/Rest_Armin.pdf) by A. Rest), first illustrated by [Couderc (1939)](https://ui.adsabs.harvard.edu/abs/2007ASPC..363..121S/).

```python
def echoxy0(t, c, Ntheta=1000, foci=False):

    a = 0.5 * (t + np.sqrt(t**2 + 4 * (c**2)))
    b = np.sqrt(0.5 * (t**2 + np.sqrt(t**4 + 4 * t**2 * c**2)))
    
    if foci:
        foci_length = np.sqrt(np.abs(a**2 - b**2))
        
        if a>=b:
            fx = np.array([-foci_length, foci_length])
            fy = np.array([0.,0.])
        if a<b:
            fx = np.array([0.,0.])
            fy = np.array([-foci_length, foci_length])
        return fx,fy

    else:
        theta = np.arange(0, 2*np.pi + 2*np.pi / Ntheta, 2*np.pi / Ntheta) # the angle about the origin

        x = a * np.cos(theta)
        y = b * np.sin(theta)
        return x,y
```

```python
# or we can sample using very simple gaussian uncertainties

Nsp = 50

R0 = 169000 # distance to the SNe event
dR0 = 10450 # distance error I guess, 6%

c = R0/2
dc = dR0/2

t = 25000 
dt = 0.5 # assume event was 6-mo long, use as uncertainty


plt.figure(figsize=(4,4))

for k in range(Nsp):
    x,y = echoxy0(np.random.normal(loc=t, scale=dt), np.random.normal(loc=c, scale=dc))
    plt.plot(x/1e4,y/1e4, c='k', alpha=0.1,lw=1)
    
fx,fy = echoxy0(t,c, foci=True)
plt.scatter(fx/1e4,fy/1e4)

plt.xlim(-12,12)
plt.ylim(-12,12)

# plt.savefig(' .pdf', dpi=150, bbox_inches='tight', pad_inches=0.25)
```

```python
# That doesn't look physically correct, b/c the error is not the total distance R0, 
# but the distance TO the 2nd foci (SN event)

Nsp = 100

fx,fy = echoxy0(t, c, foci=True)

plt.figure(figsize=(4,4))

for k in range(Nsp):
    c_k = np.random.normal(loc=c, scale=dc)
    x, y = echoxy0(np.random.normal(loc=t, scale=dt), c_k)
    fx_k, fy_k = echoxy0(t, c_k, foci=True)
    plt.plot((x - fx_k[1])/1e4, y/1e4, c='k', alpha=0.1,lw=1)
    
fx,fy = echoxy0(t, c, foci=True)
plt.scatter((fx - fx[1])/1e4, fy/1e4)

plt.xlim(-22,2)
plt.ylim(-12,12)

plt.savefig('d_error.png', dpi=150, bbox_inches='tight', pad_inches=0.25)
```

```python
t = 33. # now set to current age, 33yrs

x,y = echoxy0(t, c, Ntheta=10000)
fx,fy = echoxy0(t, c, foci=True)


plt.figure(figsize=(4,6))
plt.plot(x-fx[1],y, c='k')
for k in range(0,100,3):
    x,y = echoxy0(t+k, c, Ntheta=10000)
    plt.plot(x-fx[1],y, c='k', alpha=1/(k+1))


plt.scatter(fx-fx[1],fy)

plt.xlim(-75,25)
plt.ylim(-75,75)
plt.xlabel('Dist from Earth (ly)')
plt.savefig('today.png', dpi=150, bbox_inches='tight', pad_inches=0.25)
```
